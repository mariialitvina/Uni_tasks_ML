## Outliers

```python
def outliers(df):
    num_var = list(df._get_numeric_data().columns)
    for col_names in num_var:
        df[col_names] = df[col_names].apply(lambda y: df[col_names].mean()-3*df[col_names].std() 
                            if y < df[col_names].mean()-3*df[col_names].std() else y)
        df[col_names] = df[col_names].apply(lambda y: df[col_names].mean()+3*df[col_names].std() 
                            if y > df[col_names].mean()+3*df[col_names].std() else y)
    return(df)

df = outliers(df)
df.describe()
```

## Regression Neural Network

```python
# Importing prepared dataset
import pickle
input = open('data_kr1.pkl', 'rb')
obj = pickle.load(input)
input.close()
X_train = obj["X_train"]
X_test = obj["X_test"]
y_train = obj["y_train"]
y_test = obj["y_test"]

y_train = np.asarray(y_train).reshape(-1,1)
y_test = np.asarray(y_test).reshape(-1,1)
```

```python
X_train.head(5)
```

Для нейронных сетей необходимо шкалирование данных.

```python
# Feature Scaling
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler().fit(X_train)
X_train = sc_X.transform(X_train)
X_test = sc_X.transform(X_test)
sc_y = StandardScaler().fit(y_train)
y_train = sc_y.transform(y_train)
y_test = sc_y.transform(y_test)
```

```python
# Importing the Keras libraries and packages
import keras
from keras.models import Sequential #сети с последовательной передачей сигнала
from keras.layers import Dense  #для описания слоев нейронной сети
```

```python
# Initialising the ANN
rnn = Sequential()

# Adding the input layer and the first hidden layer
rnn.add(Dense(5, activation = 'tanh', input_dim = 4)) #на вход 4 переменных

# Adding the second hidden layer
rnn.add(Dense(4, activation = 'tanh'))

# Adding the output layer
rnn.add(Dense(1, activation = 'linear'))

# Compiling the ANN
rnn.compile(optimizer='adam', loss='mean_squared_error', metrics = ['accuracy'])
```

```python
# Fitting the ANN to the Training set
rnn.fit(X_train, y_train, batch_size = 8, epochs = 100)
```

```python
# Predicting the Test set results
y_pred = rnn.predict(X_test)
```

```python
# Inversing scaling
from sklearn.preprocessing import StandardScaler
y_hat = sc_y.inverse_transform(y_pred)
y_test = sc_y.inverse_transform(y_test)
X_test = sc_X.inverse_transform(X_test)
```

```python
from math import sqrt
from sklearn.metrics import mean_squared_error
sqrt(mean_squared_error(sc_y.inverse_transform(y_train), sc_y.inverse_transform(rnn.predict(X_train)))), sqrt(mean_squared_error(y_test, y_hat))
```

В сравнении с многофакторной моделью, ошибки стали меньше, что на обучающей, что на тестовой выборке. Сеть переобучена немного, но это же было и в прошлой модели.

```python
# Visualising the Test set results
plt.scatter(X_test[:,3], y_test, color = 'red')
plt.plot(X_test[:,3], y_hat, 'bo')
plt.title('Торговельна площа, кв. м. vs Количество чеков')
plt.xlabel('Торговельна площа, кв. м.')
plt.ylabel('Количество чеков')
plt.show()
```
