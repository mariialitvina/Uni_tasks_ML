## Missing data

Думаю, что пропуски в полях лучше заменить на моду в Количество чеков и Средний чек. Для 'Населення' и Середній дохід населення лучше использовать среднее значение.

```python
from sklearn.impute import SimpleImputer

'''SimpleImputer - класс
imputer_mean - экземпляр
fit - метод класса'''
#numeric
imputer_mean = SimpleImputer(missing_values=np.nan, strategy='most_frequent')
df[['Количество чеков']] = imputer_mean.fit_transform(df[['Количество чеков']]).round()

imputer_mean = SimpleImputer(missing_values=np.nan, strategy='most_frequent')
df[['Средний чек']] = imputer_mean.fit_transform(df[['Средний чек']]).round()

imputer_mean = SimpleImputer(missing_values=np.nan, strategy='mean')
df[['Населення']] = imputer_mean.fit_transform(df[['Населення']]).round()

imputer_mean = SimpleImputer(missing_values=np.nan, strategy='mean')
df[['Середній дохід населення']] = imputer_mean.fit_transform(df[['Середній дохід населення']]).round()
```

## Outliers

```python
def outliers(df):
    num_var = list(df._get_numeric_data().columns)
    for col_names in num_var:
        df[col_names] = df[col_names].apply(lambda y: df[col_names].mean()-3*df[col_names].std() 
                            if y < df[col_names].mean()-3*df[col_names].std() else y)
        df[col_names] = df[col_names].apply(lambda y: df[col_names].mean()+3*df[col_names].std() 
                            if y > df[col_names].mean()+3*df[col_names].std() else y)
    return(df)

df = outliers(df)
df.describe()
```

## K-Means Clustering

Продолжаю с теми же данными

```python
df
```

```python
# Using the elbow method to find the optimal number of clusters
from sklearn.cluster import KMeans
wcss = []
for i in range(1, 15):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 0)
    kmeans.fit(df)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 15), wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

Самое большое падение на 3х кластерах, но возможно также рассмотреть деление на 6 и 3 класстеров.

```python
# Using the silhouette method to find the optimal number of clusters
from sklearn.metrics import silhouette_score

sil = []

for k in range(2, 15):
  kmeans = KMeans(n_clusters = k).fit(df)  
  preds = kmeans.fit_predict(df)
  sil.append(silhouette_score(df, preds, metric = 'euclidean'))


plt.plot(range(2, 15), sil)
plt.title('Silhouette Method')
plt.xlabel('Number of clusters')
plt.ylabel('Sil')
plt.show()

for i in range(len(sil)):
 print(str(i+2) +":"+ str(sil[i]))
```

Исходя из 2х методов - Локтя и Силуета, а также малого кол-ва наблюдений, предпочту разделить на 3 кластеров.

```python
# Fitting K-Means to the dataset
km = KMeans(n_clusters = 3, init = 'k-means++', random_state = 0).fit_predict(df)
kms = KMeans(n_clusters = 3, random_state = 0).fit(df)
pd.DataFrame(kms.cluster_centers_, columns = ['Успех','Торговельна площа, кв. м.','Количество касс','Тип','Рейтинг в гуглі','Количество чеков',
                                             'Средний чек','Населення','Середній дохід населення',
                                 'Супермаркет','Магазин біля дому']).round(2)
```

Есть магазины не успешные, среднячи и успешные. 

У успешных как правило больше площадь, они новые, хороший рейтинг, много чеков=продаж, достаточно большой средний чек и, видимо, больше 1й кассы.

Есть не успешные магазины - это зачастую маленькие, старые магазины с плохими отзывами.

Магазины средней успешности также с маленькой торговой площадью, но за счет, видимо, того, что они новые это позволяет получать хорошие отзывы.
